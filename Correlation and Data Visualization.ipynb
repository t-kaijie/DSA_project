```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
Data = pd.read_csv('train_values.csv')
Data.head()
#importing the "train_values.csv" data that was downloaded
```

```python
Labels = pd.read_csv('train_labels.csv')
Labels.head()
#importing the "train_labels.csv" data that was downloaded
```

# Correlation

```python
predictorlist = pd.DataFrame(Data[["age","count_floors_pre_eq","height_percentage","area_percentage"]])

damage_grade = pd.DataFrame(Labels["damage_grade"]) 

correlation=[]
alpha_value = []
weight_given = []

for var in predictorlist:
    predictor = pd.DataFrame(Data[var])      
    correlation.append((Data[var]).corr(Labels["damage_grade"]))
    correlation.append(var)
    alpha_value.append((Data[var]).corr(Labels["damage_grade"]))

print(correlation)
sum_of_weight = sum(alpha_value)
print(sum_of_weight)
for position in range(len(alpha_value)):
    weight_given.append((alpha_value[position] / sum_of_weight).round(5))
    
a_1, a_2, a_3, a_4 = weight_given[0], weight_given[1], weight_given[2], weight_given[3]


#Values in the correlation list are the correlation value for age, count_floors_pre_eq, height_percentage and area_percentage respectively.  
```

```python
#Weighted Correlation coefficient, alpha

correlation_coefficient_list = []
for i in correlation:
    correlation_coefficient = i/(sum(correlation))
    correlation_coefficient_list.append(correlation_coefficient)

print(correlation_coefficient_list)
#Values in the correlation_coefficient_list are the values of alpha for age, count_floors_pre_eq, height_percentage and area_percentage respectively.   
```

# Data Visualization

```python
# Create a joint dataframe by concatenating the two variables
jointDF = pd.concat([predictorlist, damage_grade], axis = 1)

# Calculate the correlation between the two columns/variables
sb.heatmap(jointDF.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```
